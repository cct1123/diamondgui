```python
import time
import sys
import os
path_project = "\\".join(os.getcwd().split("\\")[:-1])
# caution: path[0] is reserved for script path (or '' in REPL)
sys.path.insert(1, path_project)

from calibration.thzreflection_trace import THzReflectionTrace
import matplotlib.pyplot as plt

# from nspyre import InstrumentGateway
# from nspyre import DataSource
# from rpyc.utils.classic import obtain
# gw = InstrumentGateway(addr='127.0.0.1')
# ds = DataSource(thzref._name, addr='127.0.0.1')
# gw.connect()
# gw.mwsyn.open()
# gw.mwsyn.cw_frequency(16.0)
# gw.mwsyn.close()

thzref = THzReflectionTrace()
print("Measurement Name: ", thzref._name)

mwfreq = 395 # [GHz] MW frequency after AMC
mwpower = 5.0 #[V]
pulse_rate = 0.3124E6#[Hz], better to be multiple of the daq sampling rate
daq_max = 0.02 #[V]
daq_min = -0.02 #[V]
daq_srate = 500E3 #[Hz]
refresh = 20 #[Hz]
window = 2.0#[s]

thzref.set_paraset(mwfreq=mwfreq, 
                   mwpower=mwpower, 
                   pulse_rate=pulse_rate, 
                   daq_max=daq_max, 
                   daq_min=daq_min, 
                   daq_srate=daq_srate, 
                   refresh=refresh, 
                   window=window)

thzref.set_stoptime(window+5)

# thzref.start();
```

```python
thzref.start();
```
