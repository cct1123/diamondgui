```python
import numpy as np
import matplotlib.pyplot as plt
import pickle
```

```python
file_path = 'output\\20240429_unknow MW pol\\fullfreqsweep_(freq_signal_background)_20240429.txt'

columns_to_load = (0, 1, 2)  # First and third columns
selected_data = np.loadtxt(file_path, usecols=columns_to_load)
freq_actual = selected_data[:, 0]
signal_av = selected_data[:, 1]
background_av = selected_data[:, 2]
dark_pl = 0.219/1000.0
contrast = (signal_av-background_av)/(background_av+dark_pl)

fig, ax1 = plt.subplots(figsize=(6, 3))  # Adjust the figsize here
idx = np.argmin(contrast)
print(f"Guess NV transition: {freq_actual[idx]}GHz")
ax1.plot(freq_actual, contrast*100.0, linewidth=2, label="contrast", color='royalblue')

ax1.set_xlabel("Frequency [GHz]")
ax1.set_ylabel("Contrast [%]", color='royalblue')
# ax1.legend(loc='lower left')
plt.title("ODMR (full range)")
plt.show()
```
